```
import torch 
import torchvision
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
import numpy as np
import torchvision.transforms as transforms
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

# 活性化関数

## ELU

```
x = torch.linspace(-6, 6, 200, dtype=torch.float)
m = nn.ELU()
y = m(x)
plt.plot(x.numpy(), y.numpy(), label='')
plt.ylim((-6, 6))
```

## LeakyReLU

```
x = torch.linspace(-6, 6, 200, dtype=torch.float)
m = nn.LeakyReLU(0.1)
y = m(x)
plt.plot(x.numpy(), y.numpy(), label='')
plt.ylim((-6, 6))
```

## LogSigmoid

```
x = torch.linspace(-6, 6, 200, dtype=torch.float)
m = nn.LogSigmoid()
y = m(x)
plt.plot(x.numpy(), y.numpy(), label='')
plt.ylim((-6, 6))
```

## PReLU

```
x = torch.linspace(-6, 6, 200, dtype=torch.float)
m = nn.PReLU()
y = m(x)
plt.plot(x.data.numpy(), y.data.numpy(), label='')
plt.ylim((-6, 6))
```

## ReLU

```
x = torch.linspace(-6, 6, 200, dtype=torch.float)
m = nn.ReLU()
y = m(x)
plt.plot(x.numpy(), y.numpy(), label='')
plt.ylim((-6, 6))
```

## ReLU6

```
x = torch.linspace(-7, 7, 200, dtype=torch.float)
m = nn.ReLU6()
y = m(x)
plt.plot(x.numpy(), y.numpy(), label='')
plt.ylim((-7, 7))
```

## SELU

```
x = torch.linspace(-6, 6, 200, dtype=torch.float)
m = nn.SELU()
y = m(x)
plt.plot(x.numpy(), y.numpy(), label='')
plt.ylim((-6, 6))
```

## Sigmoid

```
x = torch.linspace(-10, 10, 200, dtype=torch.float)
m = nn.Sigmoid()
y = m(x)
plt.plot(x.numpy(), y.numpy(), label='')
plt.ylim((0, 1))
```

## Softplus


```
x = torch.linspace(-6, 6, 200, dtype=torch.float)
m = nn.Softplus()
y = m(x)
plt.plot(x.numpy(), y.numpy(), label='')
plt.ylim((-6, 6))
```

```
#betaを0.4に
x = torch.linspace(-6, 6, 200, dtype=torch.float)
m = nn.Softplus(beta=0.4)
y = m(x)
plt.plot(x.numpy(), y.numpy(), label='')
plt.ylim((-6, 6))
```

```
#betaを2に
x = torch.linspace(-6, 6, 200, dtype=torch.float)
m = nn.Softplus(beta=2)
y = m(x)
plt.plot(x.numpy(), y.numpy(), label='')
plt.ylim((-6, 6))
```

##Tanh



```
x = torch.linspace(-6, 6, 200, dtype=torch.float)
m = nn.Tanh()
y = m(x)
plt.plot(x.numpy(), y.numpy(), label='')
plt.ylim((-1, 1))
```

## Softmax

```
x = torch.linspace(-6, 6, 200, dtype=torch.float)
m = nn.Softmax(dim=0)
y = m(x)
plt.plot(x.numpy(), y.numpy(), label='')
```

## LogSoftmax

```
x = torch.linspace(0, 1, 200, dtype=torch.float)
m = nn.LogSoftmax(dim=0)
y = m(x)
```
