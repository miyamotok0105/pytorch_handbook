# 11.1 torchvision.datasetsパッケージ

```python
%matplotlib inline
from matplotlib import pyplot as plt
import torchvision.datasets as dset
```

# MNIST

```python
mnistdata = dset.MNIST(root='./dataset/mnist_root', download=True)

plt.subplot(131)
plt.imshow(mnistdata[0][0].convert('RGB'))
plt.subplot(132)
plt.imshow(mnistdata[1][0].convert('RGB'))
plt.subplot(133)
plt.imshow(mnistdata[2][0].convert('RGB'))

plt.show()
```
