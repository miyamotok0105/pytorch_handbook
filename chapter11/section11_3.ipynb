# 11.3 torchvision.transformsパッケージ

```python
%matplotlib inline
from matplotlib import pyplot as plt
import torchvision.datasets as dset
import torchvision.transforms as transforms

stl10data = dset.STL10(root='./dataset/stl10_root', download=True)
```

## torchvision.transforms

```python
# データ前処理の構成を定義する
tf = transforms.Compose([transforms.Resize(64),
                         transforms.ToTensor(),
                         transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))])
```

## transforms.RandomApply

```python
# 前処理を行うかをランダムに決める
tf = transforms.Compose([transforms.RandomApply([transforms.CenterCrop(32),
                                                 transforms.Resize(64)], p=0.4),
                         transforms.ToTensor()])
```

## transforms.RandomChoice

```python
# ランダムに前処理を選択する
tf = transforms.Compose([transforms.RandomChoice([transforms.CenterCrop(32),
                                                  transforms.Resize(32)]),
                         transforms.ToTensor()])
```

## transforms.RandomOrder

```python
# ランダムな順序で前処理を実行する
tf =  transforms.Compose([transforms.RandomOrder([transforms.CenterCrop(32), 
                                                  transforms.Resize(32)]),
                          transforms.ToTensor()])
```

## transforms.CenterCrop

```python
img_0 = stl10data[7][0]

ccrop_1 = transforms.CenterCrop(64)
img_1 = ccrop_1(img_0)

ccrop_2 = transforms.CenterCrop(32)
img_2 = ccrop_2(img_0)

plt.subplot(131)
plt.imshow(img_0)
plt.subplot(132)
plt.imshow(img_1)
plt.subplot(133)
plt.imshow(img_2)

plt.show()
```

## transforms.RandomCrop

```python
img_0 = stl10data[16][0]

rcrop_1 = transforms.RandomCrop(96, padding=32) #, padding_mode='constant')
img_1 = rcrop_1(img_0)

rcrop_2 = transforms.RandomCrop(128, padding=0, pad_if_needed=True)
img_2 = rcrop_2(img_0)

plt.subplot(131)
plt.imshow(img_0)
plt.subplot(132)
plt.imshow(img_1)
plt.subplot(133)
plt.imshow(img_2)

plt.show()
```

## transforms.RandomResizedCrop

```python
img_0 = stl10data[45][0]

rrcrop_1 = transforms.RandomResizedCrop(64, scale=(0.1, 0.2), ratio=(2., 3.), interpolation=2)
img_1 = rrcrop_1(img_0)

rrcrop_2 = transforms.RandomResizedCrop(64, scale=(0.5, 0.6), ratio=(0.1, 0.2), interpolation=2)
img_2 = rrcrop_2(img_0)

plt.subplot(131)
plt.imshow(img_0)
plt.subplot(132)
plt.imshow(img_1)
plt.subplot(133)
plt.imshow(img_2)

plt.show()
```

## transforms.Resize

```python
img_0 = stl10data[107][0]

rsize_1 = transforms.Resize((32, 64))
img_1 = rrcrop_1(img_0)

rsize_2 = transforms.Resize((64, 32))
img_2 = rrcrop_2(img_0)

plt.subplot(131)
plt.imshow(img_0)
plt.subplot(132)
plt.imshow(img_1)
plt.subplot(133)
plt.imshow(img_2)

plt.show()
```

## transforms.Pad

```python
img_0 = stl10data[109][0]

pad_1 = transforms.Pad((4, 8, 16, 32), fill=(0, 64, 128), padding_mode='constant')
img_1 = pad_1(img_0)

pad_2 = transforms.Pad((4, 8, 16, 32), padding_mode='edge')
img_2 = pad_2(img_0)

pad_3 = transforms.Pad((4, 8, 16, 32), padding_mode='reflect')
img_3 = pad_3(img_0)

pad_4 = transforms.Pad((4, 8, 16, 32), padding_mode='symmetric')
img_4 = pad_4(img_0)

plt.subplot(151)
plt.imshow(img_0)
plt.subplot(152)
plt.imshow(img_1)
plt.subplot(153)
plt.imshow(img_2)
plt.subplot(154)
plt.imshow(img_3)
plt.subplot(155)
plt.imshow(img_4)

plt.show()
```

## transforms.ColorJitter

```python
img_0 = stl10data[112][0]

cjit_1 = transforms.ColorJitter(brightness=0.5, contrast=0, saturation=0, hue=0)
img_1 = cjit_1(img_0)

cjit_2 = transforms.ColorJitter(brightness=0, contrast=0.5, saturation=0, hue=0)
img_2 = cjit_2(img_0)

cjit_3 = transforms.ColorJitter(brightness=0, contrast=0, saturation=0.5, hue=0)
img_3 = cjit_3(img_0)

cjit_4 = transforms.ColorJitter(brightness=0, contrast=0, saturation=0, hue=0.5)
img_4 = cjit_4(img_0)

plt.subplot(151)
plt.imshow(img_0)
plt.subplot(152)
plt.imshow(img_1)
plt.subplot(153)
plt.imshow(img_2)
plt.subplot(154)
plt.imshow(img_3)
plt.subplot(155)
plt.imshow(img_4)

plt.show()
```

## transforms.Grayscale

```python
img_0 = stl10data[116][0]

gray_1 = transforms.Grayscale(3)
img_1 = gray_1(img_0)

plt.subplot(121)
plt.imshow(img_0)
plt.subplot(122)
plt.imshow(img_1)

plt.show()
```

## transforms.RandomVerticalFlip

## transforms.RandomRotation

```python
img_0 = stl10data[137][0]

rot_1 = transforms.RandomRotation(45)
img_1 = rot_1(img_0)

rot_2 = transforms.RandomRotation((50, 60), expand=True)
img_2 = rot_2(img_0)

rot_3 = transforms.RandomRotation((50, 60), center=(0, 0))
img_3 = rot_3(img_0)

plt.subplot(141)
plt.imshow(img_0)
plt.subplot(142)
plt.imshow(img_1)
plt.subplot(143)
plt.imshow(img_2)
plt.subplot(144)
plt.imshow(img_3)

plt.show()
```
